```python
def load_image(filename, size=(256,256)):
    # load image with the preferred size
    pixels = load_img(filename, target_size=size)
    # convert to numpy array
    pixels = img_to_array(pixels)
    # scale from [0,255] to [-1,1]
    pixels = (pixels - 127.5) / 127.5
    # reshape to 1 sample
    pixels = expand_dims(pixels, 0)
    return pixels
```

```python
model = load_model('Shoes_Checkpoints/model_aug_ck_0051000.h5')
```

```python
img = 'invertedimg2.jpg'
```

```python
#img = path + i
#ext = '.jpg.jpg'
#img = img + ext
f = pyplot.figure()

f.add_subplot(2,2, 1)
src_image = load_image(img)
pyplot.imshow(src_image[0])
pyplot.axis('off')


pyplot.axis('off')
#model = load_model('gen_model/model_0018000.h5')
gen_image = model.predict(src_image)
gen_image = (gen_image + 1) / 2.0
f.add_subplot(2,2, 2)
pyplot.imshow(gen_image[0])
pyplot.imsave('new_gen2.jpg', gen_image[0])
pyplot.axis('off')
```

```python
from PIL import Image
import os
import PIL
import cv2
import numpy as np
from matplotlib import pyplot
```

```python
img = Image.open('img6.jpeg')
print(img.size)
```

```python
img = img.resize((256,256))
print(img.size)
```

```python
img
```

```python
img = np.array(img)
```

```python
canny = cv2.Canny(np.array(img), 120, 340)
```

```python
pyplot.subplot(1, 2, 1)
pyplot.imshow(canny, 'gray')
pyplot.show()
```

```python
invert = cv2.bitwise_not(canny)
```

```python
cv2.imwrite('invertedimg2.jpg', invert)
```

```python
cv2.imshow('invert', invert)
#cv2.imwrite('inverted', invert)
cv2.waitKey(0) 
#closing all open windows 
cv2.destroyAllWindows() 
```
